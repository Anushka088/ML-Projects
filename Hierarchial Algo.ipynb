```python
dataset=pd.read_csv('Logistic_Iris.csv')
dataset.head
```

```python
x=dataset.iloc[:,[0,1,2,3]].values
x
```

```python
sns.pairplot(dataset)
```

```python
Elbow=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(x)
    Elbow.append(kmeans.inertia_)
plt.plot(range(1,11),Elbow,marker='o')
plt.title('The Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Elbow')
plt.show()
```

```python
distance_matrix=linkage(x,method='ward',metric='euclidean')
```

```python
dn=dendrogram(distance_matrix)
plt.axhline(y=13,color='r',linestyle='--')
plt.ylim(0,15)
plt.show()
```

```python
clusters=AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
clusters.fit_predict(x)
plt.title('Agglomerative Clustering -2 Clusters')
plt.scatter(x[:,2],x[:,3], c=clusters.labels_, label=clusters.labels_)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```
