```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
x=[89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y=[21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]
```

```python
plt.scatter(x,y)
plt.xlabel("Number of Tea Drunk")
plt.ylabel("Intelligent Qoutient")
plt.title("Tea Consumption vs Intelligent")
plt.show()
```

```python
slope,intercept,r,p,std_err=stats.linregress(x,y)
```

```python
def myfunc(x):
    return slope*x+intercept
```

```python
mymodel=list(map(myfunc,x))
```

```python
print("Slope: ",slope)
print("Intercept: ",intercept)
print("Predicted Values: ",mymodel)
print("Standard Errors: ",std_err)
print("Correlation coefficient(r): ",r)
```

```python
plt.scatter(x,y)
plt.xlabel("Number of Tea Drunk")
plt.ylabel("Intelligent Qoutient")
plt.title("Regression: Tea Consumption vs Intelligent")
plt.plot(x, mymodel,lw=4,c='orange')
plt.show()
```
