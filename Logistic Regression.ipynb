```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from matplotlib.colors import ListedColormap
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
dataset=pd.read_csv("iris.csv")
dataset
```

```python
dataset.head()
```

```python
x=dataset.iloc[:,[0,1,2,3]].values
x
```

```python
y=dataset.iloc[:,4].values
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
classifier=GaussianNB()
classifier.fit(x_train,y_train)
Out[_]: GaussianNB()
```

```python
y_pred=classifier.predict(x_test)
print("Predicted Values")
print(y_pred)
```

```python
acc=accuracy_score(y_test,y_pred)*100
print("\n\nAccuracy of Navie Bayes Classifier:",acc)
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
fig,ax=plt.subplots(figsize=(6,6))
ax.imshow(cm)
ax.grid('Flase')
ax.xaxis.set(ticks=(0,1,2),ticklabels=('Predicted Setosa','Predicted Versicolor','Predicted Virginica'))
ax.yaxis.set(ticks=(0,1,2),ticklabels=('Actual Setosa','Actual Versicolor','Actual Virginica'))
ax.set_ylim(2.5,-0.5)
for i in range(3):
    for j in range(3):
        ax.text(j,i,cm[i,j],ha='center',va='center',color='white')
plt.show()
```
